```python
# Define recommendation logic based on user data

def recommend_risk_mitigation(user_data):

    alerts = []

    risk_mitigation_strategies = []

 

    # if user_data["External_Rating_Alert"] == 1:

    #     alerts.append("External Rating Alert")

    #     risk_mitigation_strategies.append("Get External Rating done from other Rating Agency.")

 

    if user_data["Regulatory_News_Alert"] == "High":

        alerts.append("Regulatory News Alert ")

        risk_mitigation_strategies.append("Seek clarity on company's strategy to react to Regulatory Changes")

   
    if user_data["Statutory_Dues_Delay"] == "High":

        alerts.append("Statutory Dues Delay ")

        risk_mitigation_strategies.append("Seek written clarification on reasons why statutory dues are delayed and submit No Dues Certificate")

  
    if user_data["Vendor_Loss_Alert"] == "High":

        alerts.append("Vendor Loss Alert")

        risk_mitigation_strategies.append("Seek clarification on reasons of loss of Vendors (Upstream/Downstream)")

 

    if user_data["Penalty_Alert"] == "High":

        alerts.append("Penalty Alert ")

        risk_mitigation_strategies.append("Seek written clarification on reasons why penalty is levied and submit No Dues Certificate")

   
    if user_data["Business_Expansion_Alert"] == "High":

        alerts.append("Business Expansion Alert ")

        risk_mitigation_strategies.append("Seek clarification on reasons of Business Expansion and Sources of Funds for the same")
 

    if user_data["Competition_Alert"] == "High":

        alerts.append("Competition_Alert ")

        risk_mitigation_strategies.append("Seek clarity on strategies to combat increasing competition")
        
 
    if user_data["Low_Market_Demand"] == "High":

        alerts.append("Low Market Demand")

        risk_mitigation_strategies.append("Submit revised plan of Marketing/Distribution of products/services")
        
    
    if user_data["Employee_Attrittion"] == "High":

        alerts.append("Employee Attrittion")

        risk_mitigation_strategies.append("Seek reasons for increasing employee attrition")
        
      
    if user_data["Exchange_Filing_Alert"] == "High":

        alerts.append("Exchange Filing Alert")

        risk_mitigation_strategies.append("Seek written clarification of Exchange filing related issues")  
        
     
    if user_data["Cross_Border_Alert"] =="High":

        alerts.append("Cross Border Alert")

        risk_mitigation_strategies.append("Submit revised Sales Strategy for Import/Export of its products")     
        
         
    if user_data["Value_Chain_Alert"] == "High":

        alerts.append("Value Chain Alert")

        risk_mitigation_strategies.append("Submit revised Sales Strategy for Import/Export of its products")    
        

    if user_data["Personal_Loans_Promoters"] == "High":

        alerts.append("Personal Loans Promoters")

        risk_mitigation_strategies.append("Demand Closure of other Loans & submit 'No Dues Certificate'.")    


    return alerts, risk_mitigation_strategies


```

```python
def alert_recommendations(input_df):
    df = input_df.copy()
    df['Alerts'] = np.nan
    df['Risk_mitigation_strategies'] = np.nan
    for index in df.index:
        user_data = df.iloc[index]
        alerts, risk_mitigation_strategies = recommend_risk_mitigation(user_data)
        if len(alerts) == 0:
            alerts.append("No alert")
            risk_mitigation_strategies.append("No risk mitigation strategies needed.")
        df['Alerts'][index] = alerts
        df['Risk_mitigation_strategies'][index] = risk_mitigation_strategies
    return df

results_df = alert_recommendations(df)
results_df
```
