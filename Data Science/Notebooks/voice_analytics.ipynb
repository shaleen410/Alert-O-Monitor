## **helper**

```python
from collections.abc import Mapping
from typing import Any, List

def chunk(xs : List[Any], size : int) -> List[List[Any]] :
    return [xs[i : i + size] for i in range(0, len(xs), size)]
```

## **restapi_helper**

```python
import requests
from typing import Dict, List

def send_post_request(uri : str, content : Dict, key : str, expected_status_codes : List[int]) -> Dict :
    headers = {"Ocp-Apim-Subscription-Key": key}
    response = requests.post(uri, headers=headers, json=content)
    if response.status_code not in expected_status_codes :
        raise Exception(f"The POST request to {uri} returned a status code {response.status_code} that was not in the expected status codes: {expected_status_codes}")
    else :
        try :
            response_json = response.json()
            return { "headers" : response.headers, "text" : response.text, "json" : response_json }
        except Exception :
            return { "headers" : response.headers, "text" : response.text, "json" : None }


def send_get_request(uri : str, key : str, expected_status_codes : List[int]) -> Dict :
    headers = {"Ocp-Apim-Subscription-Key": key}
    response = requests.get(uri, headers=headers)
    if response.status_code not in expected_status_codes :
        raise Exception(f"The GET request to {uri} returned a status code {response.status_code} that was not in the expected status codes: {expected_status_codes}")
    else :
        try :
            # response.json() throws if the response is empty.
            response_json = response.json()
            return { "headers" : response.headers, "text" : response.text, "json" : response_json }
        except Exception :
            return { "headers" : response.headers, "text" : response.text, "json" : None }

```

## **audio_transcript**

```python
# import modules
from datetime import datetime
import pendulum
from http import HTTPStatus
from os import linesep
from time import sleep
from typing import Dict, List, Tuple
import uuid
# from user_config_file import user_config
# from restapi_helper import send_post_request,send_get_request
# from helper import chunk
import pandas as pd
import uuid
from azure.storage.blob import BlobServiceClient
import io
from flashtext import KeywordProcessor

SPEECH_TRANSCRIPTION_PATH = "/speechtotext/v3.0/transcriptions"
SENTIMENT_ANALYSIS_PATH = "/language/:analyze-text"
SENTIMENT_ANALYSIS_QUERY = "?api-version=2022-05-01"
CONVERSATION_ANALYSIS_PATH = "/language/analyze-conversations/jobs"
CONVERSATION_ANALYSIS_QUERY = "?api-version=2023-11-15-preview"
CONVERSATION_SUMMARY_MODEL_VERSION = "2022-05-15-preview"

WAIT_SECONDS = 10

def create_transcription(user_config) -> str :
    uri = f"https://{user_config['speech_endpoint']}{SPEECH_TRANSCRIPTION_PATH}"
    content = {
        "contentUrls" : [user_config["input_audio_url"]],
        "properties" : {
            "diarizationEnabled" : True,
            "timeToLive" : "PT30M",
            "wordLevelTimestampsEnabled" : True,
            # "channels": [0]
        },
        "locale" : user_config["locale"],
        "displayName" : f"call_center_{datetime.now()}",
    }
    response = send_post_request(uri=uri, content=content, key=user_config["speech_subscription_key"], expected_status_codes=[HTTPStatus.CREATED])
    transcription_uri = response["json"]["self"]
    transcription_id = transcription_uri.split("/")[-1]
    # Verify the transcription ID is a valid GUID.
    try :
        uuid.UUID(transcription_id)
        return transcription_id
    except ValueError:
        raise Exception(f"Unable to parse response from Create Transcription API:{linesep}{response['text']}")


def get_transcription_status(transcription_id : str, user_config : dict) -> bool :
    uri = f"https://{user_config['speech_endpoint']}{SPEECH_TRANSCRIPTION_PATH}/{transcription_id}"
    response = send_get_request(uri=uri, key=user_config["speech_subscription_key"], expected_status_codes=[HTTPStatus.OK])
    if "failed" == response["json"]["status"].lower() :
        raise Exception(f"Unable to transcribe audio input. Response:{linesep}{response['text']}")
    else :
        return "succeeded" == response["json"]["status"].lower()


def wait_for_transcription(transcription_id : str, user_config : dict) -> None :
    done = False
    while not done :
        print(f"Waiting {WAIT_SECONDS} seconds for transcription to complete.")
        sleep(WAIT_SECONDS)
        done = get_transcription_status(transcription_id, user_config=user_config)


def get_transcription_files(transcription_id : str, user_config : dict) -> Dict :
    uri = f"https://{user_config['speech_endpoint']}{SPEECH_TRANSCRIPTION_PATH}/{transcription_id}/files"
    response = send_get_request(uri=uri, key=user_config["speech_subscription_key"], expected_status_codes=[HTTPStatus.OK])
    return response["json"]


def get_transcription_uri(transcription_files : Dict) -> str :
    # Get Transcription Files JSON response sample and schema:
    value = next(filter(lambda value: "transcription" == value["kind"].lower(), transcription_files["values"]), None) 
    if value is None :
        raise Exception (f"Unable to parse response from Get Transcription Files API:{linesep}{transcription_files['text']}")
    return value["links"]["contentUrl"]


def get_transcription(transcription_uri : str) -> Dict :
    response = send_get_request(uri=transcription_uri, key="", expected_status_codes=[HTTPStatus.OK])
    return response["json"]


def get_transcription_phrases(transcription : Dict) -> list :
    def helper(id_and_phrase : Tuple[int, Dict]) -> dict :
        (id, phrase) = id_and_phrase
        best = phrase["nBest"][0]
        speaker_number : int
        # Note: Channels are numbered from 0. Speakers are numbered from 1.
        if "speaker" in phrase :
            speaker_number = phrase["speaker"] - 1
        elif "channel" in phrase :
            speaker_number = phrase["channel"]
        else :
            raise Exception(f"nBest item contains neither channel nor speaker attribute.{linesep}{best}")
        
        TranscriptionPhrase = {
            "id" : id,
            "text" : best["display"],
            "speaker_number" : speaker_number,
            "offset" : phrase["offset"],
            "offset_in_ticks" : phrase["offsetInTicks"]
        }
        return TranscriptionPhrase
    # For stereo audio, the phrases are sorted by channel number, so resort them by offset.
    return list(map(helper, enumerate(transcription["recognizedPhrases"])))


def get_sentiments_helper(documents : List[Dict], user_config : dict) -> Dict :
    uri = f"https://{user_config['language_endpoint']}{SENTIMENT_ANALYSIS_PATH}{SENTIMENT_ANALYSIS_QUERY}"
    content = {
        "kind" : "SentimentAnalysis",
        "analysisInput" : { "documents" : documents },
    }
    response = send_post_request(uri = uri, content=content, key=user_config["language_subscription_key"], expected_status_codes=[HTTPStatus.OK])
    return response["json"]["results"]["documents"]


def get_sentiment_analysis(phrases : list, user_config : dict) ->list :
    retval= []
    # Create a map of phrase ID to phrase data so we can retrieve it later.
    phrase_data : Dict = {}
    # Convert each transcription phrase to a "document" as expected by the sentiment analysis REST API.
    # Include a counter to use as a document ID.
    documents : List[Dict] = []
    for phrase in phrases :
        phrase_data[phrase["id"]] = (phrase["speaker_number"], phrase["offset_in_ticks"])
        documents.append({
            "id" : phrase["id"],
            "language" : user_config["language"],
            "text" : phrase["text"],
        })
    # We can only analyze sentiment for 10 documents per request.
    # Get the sentiments for each chunk of documents.
    result_chunks = list(map(lambda xs : get_sentiments_helper(xs, user_config), chunk (documents, 10)))
    for result_chunk in result_chunks :
        for document in result_chunk :
            # retval.append(SentimentAnalysisResult(phrase_data[int(document["id"])][0], phrase_data[int(document["id"])][1], document))
            SentimentAnalysisResult = {
                "speaker_number" : phrase_data[int(document["id"])][0],
                "offset_in_ticks" : phrase_data[int(document["id"])][1],
                "document" : document}
            retval.append(SentimentAnalysisResult)
    return retval


def get_sentiment_confidence_scores(sentiment_analysis_results : list) -> List[Dict] :
    sorted_by_offset = sorted(sentiment_analysis_results, key=lambda x : x["offset_in_ticks"])
    return list(map(lambda result : result["document"]["confidenceScores"], sorted_by_offset))

def get_sentiment_confidence_scores_conversation(sentiment_analysis_results : list,phrases : List) -> List[Dict] :
    sorted_by_offset = sorted(sentiment_analysis_results, key=lambda x : x["offset_in_ticks"])
    sentiment_confidence_scores_json = list()
    for index in range(len(phrases)):
        sentiment_scores = {
            'id': phrases[index]['id'],
            'text': phrases[index]['text'],
            'speaker_number': phrases[index]['speaker_number'],
            'offset_in_ticks': phrases[index]['offset_in_ticks'],
            'document': {'sentiment' : sorted_by_offset[index]['document']['sentiment'], 
                        'confidenceScores' : sorted_by_offset[index]['document']['confidenceScores']},
        }
        sentiment_confidence_scores_json.append(sentiment_scores)
    return sentiment_confidence_scores_json


def transcription_phrases_to_conversation_items(phrases : List) -> List[Dict] :
    return [{
        "id" : phrase["id"],
        "text" : phrase["text"],
        "role" : "Agent" if 0 == phrase["speaker_number"] else "Customer",
        "participantId" : phrase["speaker_number"]
    } for phrase in phrases]



def request_conversation_analysis(conversation_items : List[Dict], user_config : dict) -> str :
    uri = f"https://{user_config['language_endpoint']}{CONVERSATION_ANALYSIS_PATH}{CONVERSATION_ANALYSIS_QUERY}"
    
    content = {
  "displayName": "Conversation Task Example demo",
  "analysisInput": {
    "conversations": [
      {
        "conversationItems": conversation_items,
        "modality": "text",
        "id": "conversation1",
        "language": "en"
      }
    ]
  },
  "tasks": [
    {
      "taskName": "Conversation Summarization Task",
      "kind": "ConversationalSummarizationTask",
      "parameters": {
        "summaryAspects": [
          "recap",
          "follow-up tasks"
        ]
      }
    }
  ]
}
    response = send_post_request(uri=uri, content=content, key=user_config["language_subscription_key"], expected_status_codes=[HTTPStatus.ACCEPTED])
    return response["headers"]["operation-location"]


def get_conversation_analysis_status(conversation_analysis_url : str, user_config : dict) -> bool :
    response = send_get_request(uri=conversation_analysis_url, key=user_config["language_subscription_key"], expected_status_codes=[HTTPStatus.OK])
    if "failed" == response["json"]["status"].lower() :
        raise Exception(f"Unable to analyze conversation. Response:{linesep}{response['text']}")
    else :
        return "succeeded" == response["json"]["status"].lower()


def wait_for_conversation_analysis(conversation_analysis_url : str, user_config : dict) -> None :
    done = False
    while not done :
        print(f"Waiting {WAIT_SECONDS} seconds for conversation analysis to complete.")
        sleep(WAIT_SECONDS)
        done = get_conversation_analysis_status(conversation_analysis_url, user_config=user_config)


def get_conversation_analysis(conversation_analysis_url : str, user_config : dict) -> Dict :
    response = send_get_request(uri=conversation_analysis_url, key=user_config["language_subscription_key"], expected_status_codes=[HTTPStatus.OK])
    return response["json"]
    

def conversation_transcript(recognized_phrases) -> List:
    messages=[]
    curr_sp=1
    message=""
    for i in recognized_phrases:
        if i["speaker"]==curr_sp:
            message=message+" "+i["nBest"][0]["display"]
        else:
            messages.append({"speaker":curr_sp,"message":message})
            curr_sp=i["speaker"]
            message=""+i["nBest"][0]["display"]
    return messages


def save_dataframes_to_fabric(results_df) -> None:
    sentiments_df = pd.DataFrame(columns=["AudioID", "Date", "Transcription","Sentiment"])  
    keyword_df = pd.DataFrame(columns=["AudioID", "Date", "ExtractedKeyword"])
    for index in results_df.index:
        audio_id = results_df['AudioID'][index]
        date = datetime.now().strftime("%Y-%m-%d") 
        for sentiment_details in results_df["Sentiment_scores"][index]:
            if sentiment_details["speaker_number"] == 1:
                transcription_sentence = sentiment_details["text"]
                sentiment = sentiment_details["document"]['sentiment']
                sentiments_json = {
                        "AudioID" : audio_id, 
                        "Date" : date,
                        "Transcription" : transcription_sentence,
                        "Sentiment" : sentiment
                    }
                sentiments_df = sentiments_df._append(sentiments_json, ignore_index = True)

        for extracted_keyword in  results_df["Extracted_keywords"][index]:
            keyword_json = {
                    "AudioID" : audio_id, 
                    "Date" : date,
                    "ExtractedKeyword" : extracted_keyword,
                }
            keyword_df = keyword_df._append(keyword_json, ignore_index = True)

    
    # current_datetime = datetime.now().strftime("%Y-%m-%d %H-%M-%S") 
    # str_current_datetime = str(current_datetime)
    # directory_path = "abfss://ws_alertomonitor@onelake.dfs.fabric.microsoft.com/AOM_Lakehouse.Lakehouse/Files/voiceaudiofiles"
    
    
    # voice_file_name = directory_path + "/voice_master.csv"
    # sentiment_file_name = directory_path + "/sentiment_scores.csv"    
    # voice_master_df = pd.read_csv(voice_file_name)
    # sentiment_scores_df = pd.read_csv(sentiment_file_name)
    # voice_master_df = pd.concat([voice_master_df,results_df], axis=0)
    # sentiment_scores_df = pd.concat([sentiment_scores_df,sentiments_df], axis=0)
    # results_df.to_csv(voice_file_name, index=False)
    # sentiments_df.to_csv(sentiment_file_name, index=False) 

    # voice_master_df = results_df
    # sentiment_scores_df = sentiments_df
    # voice_master_df = results_df.copy()
    convert_dict = {'Conversation' : str, 'Sentiment_scores' : str, 'Extracted_keywords' : str}
    results_df = results_df.astype(convert_dict)
    directory_path = 'abfss://ws_alertomonitor@onelake.dfs.fabric.microsoft.com/AOM_Lakehouse.Lakehouse/'
    voice_file_path = directory_path + "Tables/voice_master"
    sentiment_file_path = directory_path + "Tables/sentiment_scores"
    keyword_file_path = directory_path + "Tables/voice_keywords"
    # voice_master_df = pd.read_parquet(voice_file_path)
    # sentiment_scores_df = pd.read_parquet(sentiment_file_path)
    # voice_keywords_df = pd.read_parquet(keyword_file_path)

    # voice_master_df=spark.read.format("delta").load("Tables/voice_master")
    # voice_master_df=voice_master_df.toPandas()
    # sentiment_scores_df=spark.read.format("delta").load("Tables/voice_master")
    # sentiment_scores_df=sentiment_scores_df.toPandas()
    # voice_keywords_df=spark.read.format("delta").load("Tables/voice_master")
    # voice_keywords_df=voice_keywords_df.toPandas()

    # voice_master_df = pd.concat([voice_master_df,results_df], axis=0)
    # sentiment_scores_df = pd.concat([sentiment_scores_df,sentiments_df], axis=0)
    # voice_keywords_df = pd.concat([voice_keywords_df,keyword_df], axis=0)
    

    voice_master_df = results_df
    sentiment_scores_df = sentiments_df
    voice_keywords_df = keyword_df
    voice_master_df_write=spark.createDataFrame(voice_master_df)
    voice_master_df_write.write.format("delta").mode("overwrite").save(voice_file_path)
    sentiment_scores_df_write=spark.createDataFrame(sentiment_scores_df)
    sentiment_scores_df_write.write.format("delta").mode("overwrite").save(sentiment_file_path)
    voice_keywords_df_write=spark.createDataFrame(voice_keywords_df)
    voice_keywords_df_write.write.format("delta").mode("overwrite").save(keyword_file_path)

def save_dataframes_to_blob(user_config,results_df) -> None:
    sentiments_df = pd.DataFrame(columns=["AudioID", "Date", "Transcription","Sentiment"])  
    for index in results_df.index:
        audio_id = results_df['AudioID'][index]
        date = datetime.now().strftime("%Y-%m-%d") 
        for sentiment_details in results_df["Sentiment_scores"][index]:
            if sentiment_details["speaker_number"] == 1:
                transcription_sentence = sentiment_details["text"]
                sentiment = sentiment_details["document"]['sentiment']
                sentiments_json = {
                        "AudioID" : audio_id, 
                        "Date" : date,
                        "Transcription" : transcription_sentence,
                        "Sentiment" : sentiment
                    }
                sentiments_df = sentiments_df._append(sentiments_json, ignore_index = True)

    # Define your Azure Blob Storage account details
    account_name = user_config["account_name"]
    account_key = user_config["account_key"]
    container_name = user_config["output_container_name"]
    # Create a BlobServiceClient object
    blob_service_client = BlobServiceClient(account_url=f"https://{account_name}.blob.core.windows.net", credential=account_key)
    current_datetime = datetime.now().strftime("%Y-%m-%d %H-%M-%S") 
    str_current_datetime = str(current_datetime)

    #saving voice transcription df in blob
    voice_file_name = "voice_" + str_current_datetime+".csv"
    voice_csv_string = results_df.to_csv(index=False)
    voice_csv_bytes = str.encode(voice_csv_string)
    voice_blob_client = blob_service_client.get_blob_client(container=container_name, blob=voice_file_name)
    voice_blob_client.upload_blob(voice_csv_bytes, overwrite=True)

    #saving sentiments df in blob
    sentiment_file_name = "sentiment_" + str_current_datetime+".csv"
    sentiment_csv_string = sentiments_df.to_csv(index=False)
    sentiment_csv_bytes = str.encode(sentiment_csv_string)
    sentiment_blob_client = blob_service_client.get_blob_client(container=container_name, blob=sentiment_file_name)
    sentiment_blob_client.upload_blob(sentiment_csv_bytes, overwrite=True)



def container_file_transcription(user_config) -> None :  
    results_df = pd.DataFrame(columns=["AudioID", "Date","Avg_call_time_in_minutes","Transcription_complete","Conversation","Sentiment_scores","overall_sentiment","Audio_summary","Extracted_keywords"])  
    for input_url in user_config["input_audio_urls"]:
        audio_id = input_url.split('?')[0].split('/')[-1] + "_" + str(uuid.uuid4())
        user_config["input_audio_url"] = input_url
        transcription_id = create_transcription(user_config)
        wait_for_transcription(transcription_id, user_config)
        transcription_files = get_transcription_files(transcription_id, user_config)
        transcription_uri = get_transcription_uri(transcription_files)
        transcription_response = send_get_request(uri=transcription_uri, key="", expected_status_codes=[HTTPStatus.OK])
        transcription = transcription_response["json"]
        time_duration = transcription['duration']
        time_duration = pendulum.parse(time_duration)
        time_duration_mins = time_duration.total_minutes()
        transcription_complete = transcription['combinedRecognizedPhrases'][0]['display']
        keywordprocessor = KeywordProcessor()
        keywords = ["Profit", "great", "Growth","Bad","loss","On time", "progress", "open", 
                    "running", "well", "great", "good","Delay","shortage", "issues","shutdown", 
                    "ban", "close", "loss", "low", "resigned", "damage", "breakdown", "adverse", 
                    "protest","Happy","passion", "discipline"]
        keywordprocessor.add_keywords_from_list(keywords)
        extracted_keywords = keywordprocessor.extract_keywords(transcription_complete)
        recognized_phrases=transcription["recognizedPhrases"]
        conversation = conversation_transcript(recognized_phrases)
        transcription["recognizedPhrases"] = sorted(transcription["recognizedPhrases"], key=lambda phrase : phrase["offsetInTicks"])
        phrases = get_transcription_phrases(transcription)
        sentiment_analysis_results = get_sentiment_analysis(phrases, user_config)
        # sentiment_confidence_scores = get_sentiment_confidence_scores(sentiment_analysis_results)
        sentiment_confidence_scores_json = get_sentiment_confidence_scores_conversation(sentiment_analysis_results,phrases)
        sentiments = list()
        for sentiment_dict in sentiment_confidence_scores_json:
            sentiments.append(sentiment_dict['document']['sentiment'])
        overall_sentiment = max(set(sentiments), key = sentiments.count)
        conversation_items = transcription_phrases_to_conversation_items(phrases)
        conversation_analysis_url = request_conversation_analysis(conversation_items, user_config)
        wait_for_conversation_analysis(conversation_analysis_url, user_config)
        conversation_analysis_response = send_get_request(uri=conversation_analysis_url, key=user_config["language_subscription_key"], expected_status_codes=[HTTPStatus.OK])["json"]
        audio_summary = conversation_analysis_summary = conversation_analysis_response['tasks']['items'][0]['results']['conversations'][0]['summaries'][0]['text']
        results_json = {
            "AudioID" : audio_id, 
            "Date": datetime.now().strftime("%Y-%m-%d"),
            "Avg_call_time_in_minutes" : time_duration_mins,
            "Transcription_complete" : transcription_complete,
            "Conversation" : conversation,
            "Sentiment_scores" : sentiment_confidence_scores_json,
            "overall_sentiment" : overall_sentiment,
            "Audio_summary" : audio_summary,
            "Extracted_keywords" : extracted_keywords
        }
        results_df = results_df._append(results_json, ignore_index = True)
    # print("Dataframe : \n",results_df)

    # save_dataframes_to_blob(user_config,results_df)
    save_dataframes_to_fabric(results_df)


container_file_transcription(user_config)
```
